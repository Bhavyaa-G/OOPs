### 5. **Abstraction**
Abstraction focuses on exposing only the essential details of an object while hiding the implementation.

for example, A remote hides all the internal parts, we control using only the buttons above.

```python
from abc import ABC, abstractmethod

class Vehicle(ABC):  # Abstract Base Class
    @abstractmethod
    def move(self):
        pass

class Car(Vehicle):
    def move(self):
        print("The car is driving.")

class Bicycle(Vehicle):
    def move(self):
        print("The bicycle is being pedaled.")

# Objects of concrete classes
car = Car()
bike = Bicycle()
car.move()
bike.move()
```

- Abstract Base Classes (ABC) enforce the implementation of abstract methods in derived classes.
- The `move` method must be implemented in the child classes.
