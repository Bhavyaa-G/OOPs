### 4. **Polymorphism**
Polymorphism allows methods in different classes to have the same name but behave differently.

```python
class Bird:
    def make_sound(self):
        print("Chirp chirp!")

class Cat:
    def make_sound(self):
        print("Meow!")

# Demonstrating polymorphism
def animal_sound(animal):
    animal.make_sound()

bird = Bird()
cat = Cat()
animal_sound(bird)
animal_sound(cat)
```

- Both `Bird` and `Cat` have a `make_sound` method, but their implementations differ.
- Polymorphism enables a single interface (`animal_sound`) to handle different types of objects.
