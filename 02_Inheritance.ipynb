### 2. **Inheritance**
Inheritance allows a class (child) to inherit attributes and methods from another class (parent).

```python
class Animal:
    def __init__(self, species):
        self.species = species

    def make_sound(self):
        print("Some generic sound")

class Dog(Animal):  # Child class inheriting from Animal
    def __init__(self, breed):
        super().__init__("Dog")  # Call the parent class constructor
        self.breed = breed

    def make_sound(self):
        print("Woof woof!")

dog = Dog("Golden Retriever")
print(dog.species)  # Inherited property
dog.make_sound()  # Overridden method
```

- `super()` is used to call the parent class's constructor or methods.
- The `Dog` class overrides the `make_sound` method of the `Animal` class.

###Overriding

```python
class Employee:
    def __init__(self, name, salary):
        self.name = name
        self.salary = salary

    def display_details(self):
        print(f"Employee Name: {self.name}")
        print(f"Salary: {self.salary}")

class Manager(Employee):
    def __init__(self, name, salary, department):
        super().__init__(name, salary)
        self.department = department

    # Overriding the display_details method
    def display_details(self):
        print(f"Manager Name: {self.name}")
        print(f"Salary: {self.salary}")
        print(f"Department: {self.department}")

# Creating objects
employee = Employee("Bhavya", 50000)
manager = Manager("Bob", 120000, "AI/ML")

# Calling display_details method
print("Employee Details:")
employee.display_details()

print("\nManager Details:")
manager.display_details()
```

The `display_details` method in the 'Employee class' prints the basic details of an employee.

The 'Manager class' inherits from the 'Employee class' and overrides the display_details method to add information specific to a manager (like department).

When `manager.display_details()` is called, **the overridden method** in the 'Manager class' is executed instead of the parent class's version.
